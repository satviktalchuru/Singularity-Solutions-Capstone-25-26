```python
import pyvista as pv
import numpy as np

mesh = pv.read("pointcloud.vtk")
mesh
```

```python
n = 20000
idx = np.random.choice(mesh.n_points, n, replace=False)
sub = mesh.extract_points(idx)

sub.plot(
    scalars="c_cn",
    cmap="viridis",
    style="points",
    point_size=3,
    render_points_as_spheres=True,
)
```
